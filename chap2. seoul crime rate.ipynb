```python
station_name=[]

for name in crime_anal_police['관서명']:
    station_name.append('서울'+str(name[:-1])+'경찰서')
```

```python
station_name
```

```python
station_address=[]
station_lat=[]
station_lng=[]

for name in station_name:
    temp=gmaps.geocode(name,language='ko')
    
    temp_addr=temp[0].get("formatted_address")
    station_address.append(temp_addr)
    
    temp_geo=temp[0].get("geometry")
    station_lat.append(temp_geo['location']['lat'])
    station_lng.append(temp_geo['location']['lng'])
    
    print(name+'-->'+temp_addr)
```

```python
gu_name=[]

for addr in station_address:
    temp=addr.split()
    
    for gu in temp:
        if gu[-1]=='구':
            gu_name.append(gu)
```

```python
crime_anal_police['구별']=gu_name
```
